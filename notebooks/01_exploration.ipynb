## Chapter 1: Understanding the Data

  ### Business Context
  Every e-commerce business faces the same core challenge — understanding
  their customers. Who are the loyal ones? Who is about to leave? Who spends
  the most? Without answers to these questions, businesses make decisions
  based on gut feeling rather than data.

  In this project we are building a Customer Intelligence Platform that answers
  these questions using real transaction data from a UK-based online retailer
  covering December 2010 to December 2011.

  ### About the Dataset
  The dataset contains 541,909 transactions across 38 countries with 8 columns:
  - InvoiceNo: Unique identifier for each transaction
  - StockCode: Product identifier
  - Description: Product name
  - Quantity: Number of units purchased
  - InvoiceDate: Date and time of purchase
  - UnitPrice: Price per unit in GBP
  - CustomerID: Unique identifier for each customer
  - Country: Country where the customer is based

  ### What are we trying to find?
  Before building any model we need to understand:
  1. How clean is the data?
  2. Are there any anomalies or data quality issues?
  3. What does a typical customer look like?

### What We Found — Key Observations

  **1. Scale of the data**
  541,909 transactions is substantial real-world data. This is not a toy
  dataset — it represents an entire year of actual business operations.

  **2. The missing CustomerID problem**
  135,080 rows (25% of data) have no CustomerID. These are guest purchases —
  customers who bought without creating an account. Since all our models need
  to track individual customer behavior over time, we cannot use these rows.
  This is a common real-world data quality challenge.

  **3. Negative quantities and prices**
  The minimum Quantity is -80,995 and minimum UnitPrice is -11,062. These
  are not errors — they represent product returns and order cancellations.
  A business needs to track returns, but for our customer behavior models
  these would distort our analysis. We will remove them.

  **4. Cancelled orders**
  Some InvoiceNo values start with 'C' — these are explicitly cancelled orders.
  Including them would make customers appear to have purchased more than they
  actually did.

  ### So What?
  Raw data is never ready for modeling. Understanding these issues upfront
  means our models will be built on accurate, trustworthy data — which
  directly determines how reliable our predictions will be.
